# DATA CLEANING

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\junai\OneDrive\Desktop\Data Analyst Store\Labmentix\Instagram Analysis for Sugar Cosmetics\sugar_cosmetics_instagram_data.xlsx")
df
```

```python
# Cleaning Column Names

df.columns = (df.columns.str.strip().str.lower().str.replace(" ", "_"))
df.columns
```

```python
# Converting post_date to datetime

df['post_date'] = pd.to_datetime(df['post_date'], errors='coerce')
df
```

```python
# Ensuring numeric columns

num_cols = ['likes','comments','shares','saves','reach','impressions']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
```

```python
# Creating Calculated Fields

# Engagements
df['engagements'] = (df['likes'] + df['comments'] + df['shares'] + df['saves'])

# Engagement rate %
df['engagement_rate_pct'] = np.where(df['reach'] > 0,(df['engagements'] / df['reach']) * 100,np.nan)

# Time fields
df['month_year'] = df['post_date'].dt.to_period('M').astype(str)
df['weekday'] = df['post_date'].dt.day_name()
```

# EDA

```python
print("Shape :", df.shape)
print("\nMissing values:\n", df.isnull().sum())

print("\nPost Type Counts:\n", df['post_type'].value_counts())

print("\nNumeric Summary:\n")
df[['likes','comments','shares','saves','reach','impressions','engagements','engagement_rate_pct']].describe()
```

```python
# Extracting first hashtag 

df['primary_hashtag'] = df['hashtags'].str.split(',').str[0].str.strip()
df
```
